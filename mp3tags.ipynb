```python
import os
from enum import Enum
from dataclasses import dataclass, fields
import music_tag

# dict access returns a MetadataItem
# title_item = f['title']
# print(title_item)

# # MetadataItems keep track of multi-valued keys
# title_item.values  # -> ['440Hz']

# # A single value can be extracted
# title_item.first  # -> '440Hz'
# title_item.value  # -> '440Hz'

# # MetadataItems can also be cast to a string
# str(title_item)  # -> '440Hz'

# for key, value in f.items():
#     print(key, value)

class tags_fields(Enum):
    album = 'album'
    albumartist = 'albumartist'
    artist = 'artist'
    # artwork = 'artwork'
    comment = 'comment'
    compilation = 'compilation'
    composer = 'composer'
    discnumber = 'discnumber'
    genre = 'genre'
    lyrics = 'lyrics'
    totaldiscs = 'totaldiscs'
    totaltracks = 'totaltracks'
    tracknumber = 'tracknumber'
    tracktitle = 'tracktitle'
    year = 'year'
    # isrc = 'isrc'
    #bitrate (read only)
    #codec (read only)
    #length (read only)
    #channels (read only)
    #bitspersample (read only)
    #samplerate (read only)

@dataclass
class TAGS:
    album: str = None
    albumartist: str = None
    artist: str = None
    # artwork: str = None
    comment: str = None
    compilation: str = None
    composer: str = None
    discnumber: str = None
    genre: str = None
    lyrics: str = None
    totaldiscs: str = None
    totaltracks: str = None
    tracknumber: str = None
    tracktitle: str = None
    year: str = None
    # isrc: str = None

    @classmethod
    def get_tags(cls, path: str):
        tags = music_tag.load_file(path)
        tag_dict = {field.name: str(tags[field.name].first) for field in fields(cls)}
        return cls(**tag_dict)
    
    def set_tags(self, path: str):
        pass

path_folder = r'/Users/mbair/Documents/_BEATMAKING/PABLO PILA x GASTON - ANGEL DUST (PRE-MASTERS)/'
# path_file = r'01 - Fentanilo.mp3'

for file in os.listdir(path_folder):
    if not file.endswith('.mp3'):
        print(f'Skipping file: {file}')
        continue
    
    print(f'Processing file: {file}')
    path_file = file
    extension = os.path.splitext(path_file)[1].lower()
    # # print(extension)
    path = os.path.join(path_folder, path_file)
    f = music_tag.load_file(path)

    tags = TAGS.get_tags(path)

    tracknumber: str = path_file.split(' - ')[0].strip()
    tracktitle: str = path_file.split(' - ')[1].strip().replace(extension, '').strip()
    print(tracknumber, tracktitle)

    check = 1
    if check:
        tags.album = 'Angel dust'
        tags.albumartist = 'Pablo Pila'
        tags.artist = 'Pablo Pila, Gaston'
        # tags.artwork = 'artwork'
        tags.comment = 'Demos 2025.06.18'
        # tags.compilation = 'compilation'
        tags.composer = 'Pablo Pila'
        # tags.discnumber = '01'
        tags.genre = 'Demo'
        # tags.lyrics = 'lyrics'
        # tags.totaldiscs = 'totaldiscs'
        tags.totaltracks = '07'
        tags.tracknumber = tracknumber
        tags.tracktitle = tracktitle
        tags.year = '2025'
        # tags.isrc = 'isrc'

    # print(tags)

    for field in tags_fields:
        field_value: str = field.value
        value = getattr(tags, field_value)
        # print(field_value, value)
        if value and str(value).lower() != 'none':
            f[field_value] = str(value)
        # print(field.value)
        # print(getattr(tags, field.value))

    f.save()

    for field in tags_fields:
        v = f[field.value].first
        print(f'{field.value}: ', v)

```

```python
from gamdl.apple_music_api import AppleMusicApi
import pandas as pd

# Paso 1: extraer ID desde la URL
def extraer_playlist_id(url):
    import re
    match = re.search(r'/playlist/.+/([a-zA-Z0-9\.-]+)', url)
    return match.group(1) if match else None

def obtener_pl_list(playlist_url):
    playlist_id = extraer_playlist_id(playlist_url)

    # Paso 2: inicializar y llamar
    api = AppleMusicApi(storefront='es')
    playlist = api.get_playlist(playlist_id)

    pl_songs = []

    pl_data = playlist['relationships']['tracks']['data']
    for e in pl_data:
        attributes = e['attributes']
        print(attributes)
        url = attributes['url']
        artist = attributes['artistName']
        tracktitle = attributes['name']
        # print(artist, tracktitle, url)
        # print(f'{artist} - {tracktitle} [{url}]')

        pl_songs.append(
            {
                'artist': artist,
                'tracktitle': tracktitle,
                'url': url
            }
        )
    return pl_songs

playlist_url = 'https://music.apple.com/es/playlist/am-disco/pl.u-6mo4zrZF47kGNJ'

pl_songs = obtener_pl_list(playlist_url)
# print(pl_songs)

for track in pl_songs:
    print(track)

```

```python
from yt_dlp import YoutubeDL

def tipo_url(url: str) -> str:
    ydl_opts = {"quiet": True, "extract_flat": True}  # no descarga, solo metadatos
    with YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(url, download=False)
        if "entries" in info:
            return "playlist"
        else:
            return "video"

# playlist
url = 'https://youtube.com/playlist?list=PL4N-5yt5BWV8PBxksD9rjDiHw-rwSDTdO&si=_Uz9eRAe2iQyyXjJ'

# video
# url = r'https://youtu.be/Bd7nxZ129hU?si=SBat8A0fb2N8yyDL'

# Pruebas
print(tipo_url(url))
```

```python
from yt_dlp import YoutubeDL

def obtener_urls_playlist(playlist_url):
    ydl_opts = {
        'quiet': True,
        'extract_flat': True  # para no descargar ni procesar detalles pesados
    }
    with YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(playlist_url, download=False)
        if 'entries' not in info:
            raise ValueError("La URL no corresponde a una playlist")
        
        urls = []
        for entry in info['entries']:
            # entry puede contener 'url' o 'id' dependiendo del extractor
            video_url = entry.get('url') or entry.get('id')
            if video_url:
                # Para construir URL completa de YouTube
                if not video_url.startswith('http'):
                    video_url = f'https://www.youtube.com/watch?v={video_url}'
                urls.append(video_url)
        return urls

# Ejemplo de uso
# playlist = "https://www.youtube.com/playlist?list=PL9tY0BWXOZFvY2PFDL6DhtIE5HhlI9oKk"
url = 'https://youtube.com/playlist?list=PL4N-5yt5BWV8PBxksD9rjDiHw-rwSDTdO&si=_Uz9eRAe2iQyyXjJ'

videos = obtener_urls_playlist(url)
for url in videos:
    print(url)
```
